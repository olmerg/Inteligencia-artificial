```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

## Preprocessing

Encode the data to numbers


```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
#from transactionencoder import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary ,columns=te.columns_)
df

```

## Use apriori  Algorithm


```python
from mlxtend.frequent_patterns import apriori
#from apriori import apriori


apriori(df, min_support=0.6)
```

## Use apriori and decode the data

Her you can try [FP-MAX](http://rasbt.github.io/mlxtend/user_guide/frequent_patterns/fpmax/) or  [FP-Growth](http://rasbt.github.io/mlxtend/user_guide/frequent_patterns/fpgrowth/) 

```python
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

## Obtain the length of each group



```python
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

## Filtering the results

```python
frequent_itemsets[ (frequent_itemsets['length'] == 2) &
                   (frequent_itemsets['support'] >= 0.8) ]
```

## Making Association Rules 


```python
from mlxtend.frequent_patterns import  association_rules
#from association_rules import association_rules
result=association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)
result
```

Now you can use filtering propiertes of pandas

```python
result[ (result['confidence'] == 1) &
                   (result['support'] >= 0.8) ]
```
